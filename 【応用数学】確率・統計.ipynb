# MA10_確率・統計1
#### 確率とは二種類ある
**・頻度確率**<br>
　実験によって確かめられた確率。<br>
　頻度確率は**客観確率**ともいう。<br>
　例:「10本のうち一本だけ当 たりのクジを引いて当選する<br>
　　　確率を調べたところ10%で あった」という事実<br><br>
　※仮に実験の試行回数を増やしていくと、真の平均的な値に近づいてゆく(**大数の法則**)
 
**・ベイズ確率**<br>
　信念の度合いを数字で表したもの。<br>
　ベイズ確率は**主観確率**ともいう。<br>
　必ず信念の度合いを最大１００%とする。<br>
　信念の度合いを定めたのであれば、信念の度合い同士の演算が可能になる。<br><br>
 　例：「あなたは40%の確率でインフルエンザです」<br>
　私がインフルエンザになる確率は４０%というのは、<br>
 　ちゃんと実験したわけではないので、頻度確率になり得ない。<br>
    (⇨私が100人いるわけではない。私１００人いて、こういう症状だったら100にいたうちの40%なんて調べようがないので検証できない)<br>
　インフルエンザになりうる様々なファクターにより信念の度合いが高まる。<br>
　お医者さんの診断は、主観確率に過ぎない。<br>
 
 
#### 独立な事象の同時確率
お互いの発生には因果関係のない事象X=xと事象Y=yが、同時に 発生する確率<br>
　${𝑃(𝑋=𝑥, 𝑌=𝑦) = 𝑃(𝑋=𝑥)𝑃(𝑌=𝑦) = 𝑃(𝑌 = 𝑦, 𝑋 = 𝑥)}$<br>
 
 ※順序は問わない。順序が逆でも同じ結果になる。<br>

#### 条件確率
ある事象X=x(雨が降る確率)が与えられた下で，Y=y(交通事故に遭う確率)となる確率のことを指す。<br>
そして、XとYの積で表現することができる。<br>
例:雨が降っている条件下で交通事故に遭う確率<br>
　𝑃 (𝑌 = 𝑦 | 𝑋 = 𝑥 ) = $\frac{ P(𝑌 = 𝑦, 𝑋 = 𝑥)}{P(X=x)}$<br><br>

**雨が降っている条件下で交通事故に遭う確率に対して雨が降る確率を掛け合わせることで、同時確率を算出することができるという関係性が重要**<br><br>

余談であるが、Xとxの違いについて<br>
Xはx達全部を表す。<br>
xはXの中で１つを表す。<br>

 #### ベイズ則
 ベイズ則は、同時確率と条件確率を用いることにより、導くことができる。<br>
 **主観確率と客観確率のどちらでも使用することができる。**<br><br>
 同時確率<br>
 ${𝑃(𝑋=𝑥, 𝑌=𝑦) = 𝑃(𝑌 = 𝑦, 𝑋 = 𝑥)}$<br><br>
 条件付き確率により下記の式に変形できる<br>
 $P(𝑋 = 𝑥, 𝑌 = 𝑦)$ = $𝑃 (𝑋 = 𝑥 | 𝑌 = 𝑦)P(𝑋 = 𝑦)$　・・・ ①<br>
 $P(𝑌 = 𝑦, 𝑋 = 𝑥)$ = $𝑃 (𝑌 = 𝑦 | 𝑋 = 𝑥 )P(𝑋 = 𝑥)$　・・・ ②<br><br>
 ①と②を同時確率の式に代入するとベイズ則の式を導出できる<br><br>
 $𝑃 (𝑋 = 𝑥 | 𝑌 = 𝑦)P(𝑋 = 𝑦)$ =  $𝑃 (𝑌 = 𝑦 | 𝑋 = 𝑥 )P(𝑋 = 𝑥)$　<br>
$𝑃 (𝑋 = 𝑥 | 𝑌 = 𝑦)$ = $\frac{ 𝑃 (𝑌 = 𝑦 | 𝑋 = 𝑥 )P(𝑋 = 𝑥)}{P(𝑋 = 𝑦)}$<br>
 
 # MA12_確率・統計3
 #### 確率変数
 事象と結びつけられた数値<br>
 事象そのものを表すといっても良い<br>
 #### 確率分布
 各事象の確率が同じことがほぼなくて、確率は事象ごとによってバラバラ。<br>
 様々な事象に対する確率を表にまとめると確率がどういう分布をしているかがわかる。<br>
 事象の発生する確率の分布のことを**確率分布**という。(ある確率変数を取る分布が確率分布)<br>
 離散値であると表で表すことができる。<br>

 #### 確率分布の特徴を調べるには？
 確率分布を見ただけでは、特徴がわからない。<br>
 特徴を見出す例として**平均**がある。(分散、共分散については後述)<br>
 
 # MA13_確率・統計4
 #### 期待値( 平均 ) 
 その分布における、確率変数の「平均の値」or「ありえそうな値」<br>
 もう少し言うと確率的に出てくる値の平均値を考えるときには、期待値と呼ばれる。<br>
 **期待値**はある確率変数の変数と確率変数で表され、下記の式になる。<br><br>
 事象: 𝑋<br>
 確率変数:　$f(𝑋 = 𝑥_{k=n})$<br>
 確率: $P(𝑋 = 𝑥_{k=n})$<br><br>
 
 $E(f) = $${\Sigma_{k=1}^{n}}$$P(𝑋 = 𝑥_{k=1})$$f(𝑋 = 𝑥_{k=1})$<br><br>
 連続値の場合は、積分すること。
 
 
 # MA14_確率・統計5
期待値（平均）だけだとわからないことが多い。<br>
確率分布がものすごくバラバラな場合、期待値（平均値）通りにならない。<br>
どれだけ、期待値通りになるかが**分散**と言う考え。<br>

 #### 分散
データのちりばり具合。期待値はデータからどれだけずれているか平均化したもの。<br>
なんで、平均化するか？<br>
サンプル数が少ない場合はいいかもしれない。<br>
サンプル数が多いとズレが大きいかもしれない。（例えば５００万人もいればズレそう）<br><br>
分散の単位は、元の単位の二乗で表す。<br>
絶対値の考えもあったが場合分けが必要になるため、二乗という考えを使っている。<br><br>
分散 :
Var $f(x)　=　 E(f_{(X=x)} - E(f))^2$ <br><br>

1つのデータのバラツキに関しては、分散という考えを用いた。<br>
複数のデータのバラツキを考える場合に用いるのが、**共分散**。<br>
#### 共分散
2つのデータのバラツキとは何かというと、2つのデータ系列の違いを示すことと同義。<br>
⇨2つのデータ(X、Y)の散らばり方の違いをみる。傾向の違いを知る。<br>

正の値⇨似た傾向。似てるかも。関係性があるかも<br>
負の値⇨逆の傾向。似てないかも。無関係という意味でない。一方が増え、一方が減る関係。<br>
ゼロ　⇨関係性が乏しい。<br>
必ずわかるわけではないが、二つのデータの関係を知るために使用する。<br>

共分散 :
$Cov(f, g)　=　 E(f_{(X=x)} - E(f))E(g_{(Y=y)} - E(g))$ <br>
 
 # MA15_確率・統計6-1
**確率分布によってできることが変わってくる。**<br>
分布によって中身が変わる。一方が増えて、一方がへるなど。(確率分布がちぐはぐだったり)<br>
分散・共分散を出すのもいいけど、本当に出せるかわからない。難しいかもしれない<br>
数値計算できるかわからない。ただ単に数値計算でしか出せないのか?<br>
だから、事前に確率分布してるかを知りたい。<br>

様々ある確率分布で代表的な二つの分布を紹介する。<br>

**ベルヌーイ分布**<br>
確率変数が 0 と 1の２値の場合に使う(現象を二つに分けた場合など)<br>
頑張ればベルヌーイ分布の組み合わせで全て表すことができる<br><br>
 ・P(X=0)=1－μ <br>
 ・P(X=1))=μ <br>
 と表すと下記の式がベルヌーイ分布の式である。ただし、<br>
 $P_{(X=x)} = μ^x(1 - μ) ^{(1-x) }$<br><br>

**マルチヌーイ分布**<br>
→ 確率変数が3つ以上<br>
「カテゴリカル分布」という呼ばれるのが一般的。<br>
サイコロを転がすイメージ<br><br>
　P(X=n) = $\Pi_{k=1}^nλ_k^{[x=k]}$<br>
 
 λの[x=k]は条件式

 # MA16_確率・統計6-2
**二項分布**<br>
さっきまでは、一回やった時の話だった。N回やった場合の話に入る。<br>
 N回やった場合、それが二項分布。$\mu$に相当する部分が$\lambda$で表現されている。<br>
 ベルヌーイ分布の多試行版であり、式は下記である。<br>
 
　𝑃($x$ |  $\lambda$, $n$ ) = $\frac{ n!}{x!(n-x)}$$λ^x(1-λ)^{(n-x)}$<br><br>

組み合わせの式が入っているのは、、、<br>
例えば、表が三回といっても順序まではいわれていない。<br>
そういった場合、表が三回という事象が増える。<br>
なので、順序によって事象が増えないことを考慮するために組み合わせの式が入っている。<br>

**ガウス分布(正規分布)**<br>
今までは離散的な値を考えてきた。ガウス分布は、釣鐘型の連続分布である。<br>
中心が大きく(平均が大きく)、どんどん小さくなるグラフ。<br>

**基本的な式の形式としては、$e^{(−x^2)}$ で表現される。**<br>
**また、確率が1になるように平均・分散を用いたパラメータが付与されている。**<br>
